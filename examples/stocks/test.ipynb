```python
import pandas as pd
import numpy as np
```

```python
df: pd.DataFrame = pd.read_csv("btc_usd_data.csv")
df
```

```python
df.drop(["Dividends", "Stock Splits", "Date"], inplace=True, axis=1)
df
```

```python
sliding_window_size: int = 3
```

```python
print(len(df))
```

```python
X: list[list[int]] = []
Y: list[int] = []
for num in range(0, len(df) - sliding_window_size - 1):
    X.append([])
    for row in df[num:(num+sliding_window_size)].itertuples():
        X[num].append(row.Open)
        X[num].append(row.Close)
        X[num].append(row.High)
        X[num].append(row.Low)
        X[num].append(row.Volume)
        Y.append(df["Close"][num+sliding_window_size+1])

print(X[0])
print(Y[0])
```
